## Demonstration

Let's import the packages needed for this tutorial:

```python
import pyleoclim as pyleo
import pandas as pd
```

### Step 1: Create a `Series` object from a csv file

To do so, we will first read the data from a csv file and load into a [Pandas DataFrame](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html):

```python
df = pd.read_csv('../data/wtc_test_data_nino_even.csv')
df.head()
```

Next we create a `Series` object from the columns of the `DataFrame`:

```python
ts_nino = pyleo.Series(time =  df['t'], value = df['nino'],
                  time_name = 'Year', value_name = 'Ni√±o 3',
                  time_unit = 'CE', value_unit = '$^{\circ}$C')            
```

Let's make a simple plot. You can learn to customize this plot in the `publication_ready_figures.ipynb` tutorial:

```python
fig, ax = ts_nino.plot()
```

### Step 2: Pre-processing

Pyleoclim has multiple functionalities to pre-process a timeseries including standardizing, detrending, interpolation. You can learn about the various pre-processing steps in the `filtering_and_detrending.ipynb`, `data_imputation.ipynb`, and `basic_ts_manipulation.ipynb` tutorials. As an example of one treatment, let's standardize the data and plot it against the original values. 


```python
ts_nino_std = ts_nino.standardize()

fig, ax = ts_nino.plot(label='original', zorder=99) 
ax = ts_nino_std.plot(label='standardized', ax=ax, lgd_kwargs={'ncol': 2})
```

### Step 3: Spectral Analysis

Pyleoclim makes it simple to perform spectral analysis. Calling the `.spectral()` method will yield a `PSD` (power spectral density) object of the corresponding series. Unless specified otherwise, `.spectral()` uses Weighted Wavelet Z-transform (`method='wwz'`) can handle both evenly-spaced data and unevenly-spaced data. In some cases, it may be appropriate to specify an alternative method, like Multi-Taper Method by specifying `method='mtm'`, which for what it's worth can only handle evenly-spaced data.  

The `wwz` method is essentially a wavelet tranform algorithm and therefore much slower than the `mtm` method. You can learn more about the methods available in the `wavelet_analysis.ipynb` Tutorial.
For our example data, both methods apply.

```python
psd_wwz = ts_nino_std.spectral()  # method='wwz' by default 
psd_mtm = ts_nino_std.spectral(method='mtm')  # method='mtm' 
```

We can then plot the results of our analysis:

```python
fig, ax = psd_wwz.plot(label='WWZ',zorder=99)
ax = psd_mtm.plot(ax=ax, label='MTM')
```

To identify the periodocity of our data, we need to perform some significance test.
Currently, *Pyleoclim* supports the test using AR(1) surrogates, which can be performed by simply calling the `signif_test()` method.

```python
psd_mtm_signif = ts_nino_std.spectral(method='mtm').signif_test()
```

Calling the `plot()` method on our `psd_mtm_signif` object will return a plot of the PSD along with the AR(1) threshold curve, which we can use to identify the significant cycles.

```python
fig, ax = psd_mtm_signif.plot()
```

### Step 4: Wavelet Analysis

The current version of *Pyleoclim* supports the `wwz` method for wavelet analysis on unevenly-spaced records and the `cwt` method for evenly-spaced records. `wwz` is more flexible than `cwt` in the sense that it can be used on any record, but is much slower (no free lunch!). To perform wavelet analysis we simply call the `wavelet()` method, which will in turn create a `Scalogram` object. Since our data is evenly spaced, we'll use `cwt`.

```python
scal = ts_nino_std.wavelet(method='cwt')
```

```python
fig, ax = scal.plot(title='CWT')
```

We can perform significance analysis the same way we did when conducting spectral analysis:

```python
scal_signif = ts_nino_std.wavelet(method='cwt').signif_test()
```

In this case, regions of significance (those that exceed the 95% confidence threshold found using AR1 surrogates) will be outlined in white.

```python
fig, ax = scal_signif.plot()
```

### One more thing: method cascading

Life is short.`Pyleoclim` provides a very cool feature to make life a little bit easier: "method cascading". This feature allows us to perform a multi-step workflow in just one line.

```python
#Spectral analysis with a significance test
fig, ax = ts_nino.standardize().spectral(method='mtm').signif_test().plot()
```

```python
#Wavelet analysis with significance test
fig, ax = ts_nino.standardize().wavelet(method='cwt').signif_test().plot()
```
